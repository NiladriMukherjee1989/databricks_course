```python
%sql 
CREATE SCHEMA IF NOT EXISTS banking.gold;
```

```python
%sql 
CREATE OR REPLACE TABLE banking.gold.credit_summary AS (

with credit_risk_sum AS (
select customer_id,credit_score,risk_grade,external_active_loans
from banking.bronze.credit_bureau_reports WHERE
int(external_overdue_amount) <> 0
)

select count(customer_id) as total_customers,risk_grade 
from credit_risk_sum
group by risk_grade
)
```

```python
records_cnt=spark.sql(f"select count(*) AS CNT from banking.gold.credit_summary").collect()[0][0]

#Return the number of records inserted and it will be received by the master notebook
dbutils.notebook.exit(str(records_cnt))
```
